# Description:
This notebook downloads the following into your google drive
- 1000 AI generated images from "diffusiondb" dataset
- 1000 real images from "imagenet_v2" dataset

 The images files will be uploaded to the Github too. So don't bother with this really.

```python
!pip install datasets

import numpy as np
import cv2
from PIL import Image
from datasets import load_dataset
import tensorflow_datasets as tfds
```

# Download AI Generated Images

```python
# Load the ai images dataset (1k images)
ai_dataset = load_dataset('poloclub/diffusiondb', 'large_first_1k')
ai_dataset = ai_dataset['train']['image']
```

```python
from google.colab import drive
import os

drive.mount('/content/drive')

# Directory to save images
save_dir = '/content/drive/MyDrive/images/'

# Save each ai image
for idx, img in enumerate(ai_dataset):
    # Convert the image to a PIL Image
    pil_img = Image.fromarray(np.array(img))
    # Save the image
    pil_img.save(os.path.join(save_dir + "ai/", f'{idx}.png'))
```

# Download Real Images

```python
# Load the real images dataset (k images)
real_dataset = tfds.load('imagenet_v2/topimages', with_info=False)

real_images = []
for example in real_dataset['test']:
  image = example['image'].numpy()
  real_images.append(image)
```

```python
import random

random.seed(123)
sampled_real_images = random.sample(real_images, 1000)
```

```python
# Save each real image
for idx, img in enumerate(sampled_real_images):
    # Convert the image to a PIL Image
    pil_img = Image.fromarray(img)
    # Save the image
    pil_img.save(os.path.join(save_dir + "real/", f'{idx}.png'))
```
